```python
import pandas as pd
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sb
sb.set(color_codes=True)

data = pd.read_csv('../../kc_house_data.csv')
data.head()
```

```python
print(list(data.columns))
```

```python
reduced_columns = data.columns.difference(['id', 'date', 'waterfront', 'view', 'zipcode', 'lat', 'long'])

# against price
for rc in [rc for rc in reduced_columns if (rc != 'price')]:
    sb.regplot(x=data[rc], y=data['price'], fit_reg=False)
    plt.show()
```
